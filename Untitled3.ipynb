```python
def main():
    input_string = input("Enter a string: ")
    uppercase_string = input_string.upper()
    lowercase_string = input_string.lower()
    vowels = "aeiouAEIOU"
    vowel_count = sum(1 for char in input_string if char in vowels)
    reversed_string = input_string[::-1]
    print(f"Uppercase: {uppercase_string}")
    print(f"Lowercase: {lowercase_string}")
    print(f"Number of vowels: {vowel_count}")
    print(f"Reversed string: {reversed_string}")
if __name__ == "__main__":
    main()
```

```python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True
print("Prime numbers between 1 and 50:")
for num in range(1, 51):
    if is_prime(num):
        print(num, end=" ")
```

```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)
num = int(input("Enter a number: "))
print(f"Factorial of {num} is {factorial(num)}")
```

```python
class Car:
    def __init__(self, brand, model, year):
        self.brand = brand
        self.model = model
        self.year = year

    def display_info(self):
      print(f"Car Details: {self.year} {self.brand} {self.model}")
my_car = Car("Tesla", "model3", 2024)
my_car.display_info()
```

```python
students = {
    "Alex": 90,
    "John": 85,
    "Rob": 50
}
print("Student Scores:")
for name, score in students.items():
    print(f"{name}: {score}")
highest_score = max(students.values())
top_students = [name for name, score in students.items() if score == highest_score]
print(f"\nHighest Score: {highest_score}")
print(f"Student(s) with the highest score: {', '.join(top_students)}")
```
